```python
import pandas as pd
import sqlite3
conn = sqlite3.connect('bookstore.db')
conn.close()
```

```python
conn = sqlite3.connect('bookstore.db')
cursor = conn.cursor()
create_table_query = '''
CREATE TABLE IF NOT EXISTS Books (
    book_id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT Not Null,
    author TEXT Not Null,
    price REAL Not Null,
    stock_quantity INTEGER Default 0
    );
    '''

cursor.execute(create_table_query)

create_table_query = '''
CREATE TABLE IF NOT EXISTS Customers (
   customer_id INTEGER Primary Key Autoincrement,
   name TEXT Not Null,
   email TEXT Unique Not Null,
   city TEXT,
   join_date TEXT
);'''
cursor.execute(create_table_query)


create_table_query = '''
CREATE TABLE IF NOT EXISTS Orders (
  order_id  INTEGER Primary Key Autoincrement,
  customer_id INTEGER ,
  book_id INTEGER ,
  quantity INTEGER Not Null,
  order_date TEXT Not Null,
  total_amount REAL,
  FOREIGN KEY (customer_id) REFERENCES Customers (customer_id),
  FOREIGN KEY (book_id) REFERENCES Books (book_id)
);
'''
cursor.execute(create_table_query)


conn.commit()


query = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql(query,conn)
print(tables)

query = 'PRAGMA table_info(Books)'
columns_info = pd.read_sql(query,conn)
print(columns_info)

query = 'PRAGMA table_info(Customers)'
columns_info = pd.read_sql(query,conn)
print(columns_info)

query = 'PRAGMA table_info(Orders)'
columns_info = pd.read_sql(query,conn)
print(columns_info)


```

```python
#Find the customer who has placed the most orders (use GROUP BY and ORDER BY)
query = "SELECT customer_id,sum(quantity),sum(total_amount) FROM Orders group by customer_id order by sum(quantity) desc;"
cursor.execute(query)
order_records = cursor.fetchall()
#print(len(order_records))
print(order_records[0][0])

query = f"SELECT * FROM Customers WHERE customer_id = {order_records[0][0]} "
cursor.execute(query)
customer_toporders= cursor.fetchall()
print(customer_toporders)

```
